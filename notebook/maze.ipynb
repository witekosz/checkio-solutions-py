```python
def print_maze(maze: list, path: str = ""):
    y, x = 0, maze[0].index("S")

    moves = set()
    for move in path:
        if move == "L":
            x -= 1
        elif move == "R":
            x += 1
        elif move == "U":
            y -= 1
        elif move == "D":
            y += 1

        moves.add((y, x))
    print("Moves", moves)

    for y, row in enumerate(maze):
        for x, col in enumerate(row):
            if (y, x) in moves:
                print("+ ", end="")
            else:
                print(col + " ", end="")
        print()
```

```python
def valid(maze: list, moves):
    y, x = 0, maze[0].index("S")

    for move in moves:
        if move == "L":
            x -= 1
        elif move == "R":
            x += 1
        elif move == "U":
            y -= 1
        elif move == "D":
            y += 1

        if not(0 <= x < len(maze[0]) and 0 <= y < len(maze)):
            return False
        elif maze[y][x] == "#":
            return False

    return True
```

```python
def find_end(maze, moves):
    j, i = 0, maze[0].index("S")

    for move in moves:
        if move == "L":
            i -= 1

        elif move == "R":
            i += 1

        elif move == "U":
            j -= 1

        elif move == "D":
            j += 1

    if maze[j][i] == "X":
        print("Found: " + moves)
        print_maze(maze, moves)
        return True
    else:
        print_maze(maze, moves)
        return False
```
